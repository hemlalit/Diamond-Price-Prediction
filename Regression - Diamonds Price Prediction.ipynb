# Read Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diamonds train.csv")
```

# Basic Info & Stats

# Drop Irrelevant & Heavy Null Values Column

```python
df.drop(columns=['fluor_intensity', 'Unnamed: 0'], inplace=True)
```

```python
df.columns
```

```python
for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())  # Prints the frequency of unique values
```

```python
for col in df.columns:
    mode_value = df[col].mode()[0]
    if mode_value == "unknown":  # If mode is "unknown", use next most frequent value
        mode_value = df[col].value_counts().index[1]  # Get second most common value
    df.loc[df[col] == "unknown", col] = mode_value
```

```python
df
```

# Drop duplicated rows

```python
df = df.drop_duplicates()
```

# Replace Outliers with Upper or Lower whisker accordingly

```python
from remove_outliers import replace_outliers_with_whiskers

# Apply to specific numerical columns
num_cols = ['total_sales_price', 'carat_weight']  # Adjust column names accordingly
for col in df.columns:
    df_without_outliers = replace_outliers_with_whiskers(df, col)
```

```python
for col in df.columns:
    sns.boxplot(df_without_outliers[col])
    plt.show()
```

# Convert Categorical to Numerical Data

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()

cat_cols = df.select_dtypes(include=['object']).columns.tolist()

categories_list = [list(set(df[col].dropna())) for col in cat_cols]

df.loc[:, cat_cols] = oe.fit_transform(df[cat_cols])
df_without_outliers.loc[:, cat_cols] = oe.fit_transform(df_without_outliers[cat_cols])
```

# Split Data to X & Y

```python
X = df.drop('total_sales_price', axis=1)
y = df['total_sales_price']

X.fillna(np.mean(X['fancy_color_overtone']), axis=1, inplace=True)

X2 =  df_without_outliers.drop('total_sales_price', axis=1)
y2 = df_without_outliers['total_sales_price']

X2.fillna(np.mean(X2['fancy_color_overtone']), axis=1, inplace=True)
```

# Scale your input Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X2_scaled = scaler.fit_transform(X2)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)  #df with ouliers
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=99)  #df without ouliers
X_train_sc, X_test_sc, y_train_sc, y_test_sc = train_test_split(X_scaled, y, test_size=0.2, random_state=99) #df with ouliers
X2_train_sc, X2_test_sc, y2_train_sc, y2_test_sc = train_test_split(X2_scaled, y, test_size=0.2, random_state=99) #df without ouliers
```

```python
# X_train.fillna(np.mean(X_train['fancy_color_overtone']), axis=1, inplace=True)
X_train.isna().sum()
```

# Create regression models

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression

mse=[]

lr = LinearRegression()
lr2 = LinearRegression()
lr3 = LinearRegression()
lr4 = LinearRegression()

lr.fit(X_train, y_train)
lr2.fit(X2_train, y2_train) 
lr3.fit(X_train_sc, y_train_sc)
lr4.fit(X2_train_sc, y2_train_sc)

y_pred = lr.predict(X_test)
y2_pred = lr2.predict(X2_test)
y3_pred = lr3.predict(X_test_sc)
y4_pred = lr4.predict(X2_test_sc)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_mse =  mean_squared_error(y_test, y_pred)
lr_r2  = r2_score(y_test, y_pred)

mse.append(lr_mse)

print("MSE:", lr_mse)
print("R² Score:", lr_r2)

print("MSE:", mean_squared_error(y2_test, y2_pred))
print("R² Score:", r2_score(y2_test, y2_pred))

print("MSE:", mean_squared_error(y_test_sc, y3_pred))
print("R² Score:", r2_score(y_test_sc, y3_pred))

print("MSE:", mean_squared_error(y2_test_sc, y4_pred))
print("R² Score:", r2_score(y2_test_sc, y4_pred))
```
